```python
import numpy as np
from  matplotlib import pyplot as plt
malelist_x = []
malelist_y = []
n = 0
def function1(x): ### 함수 정의
    return 0.4

def function2(x): ### 함수 정의
    return 0.8

N = int(1e+4)  ### 랜덤 샘플링 시행 횟수 정의
W, H = 1, 1  ### 랜덤 샘플링을 할 사각형 R의 가로 세로 정의

X = np.random.random(N)   ### 각 점의 x 좌표 랜덤 샘플링
Y = np.random.random(N) ### 각 점의 y 좌표 랜덤 샘플링

F1 = function1(X) ### 각 랜덤 샘플링에 대한 f(x)를 계산
F2 = function2(X) 

in_out1 = Y < F1  ### y와 f(x) 값 비교
in_out2 = Y < F2

color = list(map(lambda x1, x2: 'dodgerblue' if x1 else ('red' if x2 else 'green'), in_out1, in_out2))
            ### 샘플링 포인트의 색깔 설정
            
malelist_x.append(X)
malelist_y.append(Y)

plt.scatter(malelist_x[n], malelist_y[n], color=color, label='A = {}'.format(np.round(4)))
            ### 샘플링 포인트 그리기

n = n + 1
            
plt.plot([0, W], [0, 0], color='black')  ### 사각형 영역 그리기
plt.plot([W, W], [0, H], color='black')
plt.plot([0, W], [H, H], color='black')
plt.plot([0, 0], [0, H], color='black')

plt.legend(loc = 'lower right') ### 그래프 그리는 옵션 추가
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
print(malelist_x, malelist_y)
```
